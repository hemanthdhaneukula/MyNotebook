```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Send a request to the URL
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table', {'class': 'historical_data_table table'})

# Check if any tables were found
if not tables:
    raise ValueError("No tables with class 'historical_data_table table' found on the page.")

# Initialize an empty DataFrame to collect the data
all_data = pd.DataFrame()

# Iterate through each table found
for table in tables:
    # Parse the table into a DataFrame
    df = pd.read_html(str(table))[0]  # Assuming only one table per page based on previous analysis
    # Append the DataFrame to the collected data
    all_data = pd.concat([all_data, df], ignore_index=True)

# Display the last five rows of the combined DataFrame
print(all_data.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Tesla stock data using yfinance
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  # Fetching one year of historical data

# Define the make_graph function
def make_graph(x_data, y_data, title=None):
    plt.figure(figsize=(10, 6))
    plt.plot(x_data, y_data, marker='o', linestyle='-')
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')
    if title:
        plt.title(title)
    plt.grid(True)
    plt.tight_layout()

# Assuming tesla_data['Date'] and tesla_data['Close'] are already defined
make_graph(tesla_data.index, tesla_data['Close'], title='Tesla Stock Prices')

# Show the plot
plt.show()
```
